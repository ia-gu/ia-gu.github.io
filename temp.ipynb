```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [0, 0, 0, 0, 0]
fig = plt.figure()
plt.ylim(0, 100)
plt.plot(x, y, label='random sampling', marker='.', color='red')
plt.plot(x, y, label='entropy sampling', marker='.', color='red')
plt.plot(x, y, label='batchbald', marker='.', color='red')
plt.plot(x, y, label='kmeans sampling', marker='.', color='red')
plt.plot(x, y, label='coreset', marker='.', color='red')
plt.plot(x, y, label='badge', marker='.', color='red')
plt.plot(x, y, label='cluster margin', marker='.', color='red')


# ラベルのみを表示する凡例を作成し、グラフ上に配置する
plt.legend(loc='upper right', fontsize=16)

# グラフを表示する
fig.savefig('bench.svg')
```
